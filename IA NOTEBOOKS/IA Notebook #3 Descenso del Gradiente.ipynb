```python
import numpy as np
import scipy as sc

import matplotlib.pyplot as plt
```

```python
# lambda es para declarar esta función como anónima
# th es un vector de parámetros por eso th[0] = x & th[1] = y 
func = lambda th: np.sin(1/2 * th[0] ** 2 - 1/4 * th[1] ** 2 + 3) * np.cos(2 * th[0] + 1 - np.e ** th[1])
```

```python
#### Generaremos los desniveles de la función (no recuerdo el nombre matemáico)####
# Número decordenadas que queremos, {resolución}
res=100
# Generar un vector de _X y _Y de 100 coordenadas entre -2 y 2
_X = np.linspace(-2,2,res)
_Y = np.linspace(-2,2,res)

# Inicializamos un array de valores que contendrán las coordenadas z de resXres
_Z = np.zeros((res, res))

# Guardaremos z para cada valor de x y y con 2 for
for ix, x in enumerate(_X):
    for iy, y in enumerate(_Y):
        _Z[iy, ix] = func([x,y])
        
# Enviamos los vectores y el array de datos, así como 100 como resolución de las líneas
plt.contourf(_X, _Y, _Z, 100)
plt.colorbar()

# Generar un punto aleatorio sobre la superficie

Theta = np.random.rand(2) * 4 - 2

_T = np.copy(Theta)
# Incremento pequeño
h= 0.001
# Radio de aprendizaje, learning ratio
lr= 0.001

# Punto inicial
plt.plot(Theta[0], Theta[1], "o", c="white")

# Vector gradiente
grad = np.zeros(2)

# Relizamos las iteraciones 
for _ in range(10000):
    
    # Realizamos las derivadas
    for it, th in enumerate(Theta):
        #Copiamos devuelta de Theta 
        _T = np.copy(Theta)
        #Agregamos el incremento h
        _T[it] = _T[it] + h
        #Derivamos
        deriv = (func(_T) - func(Theta)) / h
        #Guardamos el vector gradiente 
        grad[it] = deriv
    #Ajustamos el ratio de aprendizaje al vector gradiente y se lo restamos a el vector Theta
    Theta = Theta - lr * grad
    #Cada 100 iteraciones que se imprima un punto para seguir el proceso
    if(_ % 100 == 0):
        plt.plot(Theta[0], Theta[1], ".", c="red")

# Imprimimos el punto aleatorio
plt.plot(Theta[0], Theta[1], "o", c="green")

plt.show()
```
