```python
#Añadimos la librería para IA de google, tensorflow
import tensorflow as tf
import keras

#Importamos los utilitarios.
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Importamos la librería de trabajo
from keras.applications.inception_v3 import InceptionV3, decode_predictions
from keras import backend as k

#Aquí importamos la librería keras para procesamiento de imagenes.
from keras.preprocessing import image
```

```python
Iv3 = InceptionV3()
```

```python
print(Iv3.summary())
```

```python
#Cargamos la imagen, la redimensionamos en 299x299 y, por último,
#transdormamos la imagen a un matriz
x = image.img_to_array(image.load_img("./IA2/perro.jpg", target_size=(299, 299)))

#Cambiamos de rango, pasamos de  {0}-{255} -a {-1}-{1}

x /= 255
x -= 0.5
x *= 2

x = x.reshape(1, x.shape[0], x.shape[1], x.shape[2])

Y = Iv3.predict(x)

decode_predictions(Y)
```

#Ataques adversarios

```python
#punto de entrada de datos.
inp_layer = Iv3.layers[0].input
#último elemento de un array como -1 (última capa)
out_layer = Iv3.layers[-1].output

#Es la clase del limón.
target_class = 951

#Función de coste, donde 
# lo que queremos maximizar es la clase target.
loss = out_layer[0, target_class]

#Calculamos el array de gradientes de la función de
# coste respecto a la imagen de entrada.
# [0] es para inicializar el array
grad = k.gradients(loss, inp_layer)[0]

#Función que se ejecutará iterativamente para que se optimice
# k.function (), le damos los valores de entrada y los que esperamos de
# salida.***
optimaze_gradiant = k.function([inp_layer, k.learning_phase()], [grad, loss])

#Clonamos una variable igual a X (gatito)
adv = np.copy(x)

#Perturbación es para que mantenga en un rango de valores que pueda manipular. 
#Ese MAX and MIN lo llamaremos umbral.
pert = 0.01

max_pert = x + 0.01
min_pert = x - 0.01

#inicializmos coste
cost = 0.0

#bucle iterativo donde ejecutaremos la optimización
# hasta que alcanze el 95% de optimización.
while cost < 0.95:
    #guardamos en gr el gradiente de cada iteración y el coste,
    # y enviamos como entrada adv y 0, este es para
    # declarar que es de entrenamiento
    gr, cost = optimaze_gradiant([adv, 0])
    
    # añadimos a nuestra imagen el gradiente
    adv += gr
    
    # np.clip saturá una imagen entre un max y min de rango
    adv = np.clip(adv, min_pert, max_pert)
    adv = np.clip(adv, -1, 1)
    
    print("Costo es :", cost)

hacked = np.copy(adv)
```

```python
#Realizamos las operaciones para que ...
adv /= 2
adv += 0.5
adv *= 255
```
